```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.arange(-10, 10, 0.3)[:32].reshape((32, 1))
y = -6 * x + 0.3 * np.random.normal(loc=0.0, scale=30.0, size=x.shape)
reg = LinearRegression()
reg.fit(x,y)
y_predict = reg.predict(x)

#绘图并输出方程、系数
print("Equation:","y=%.4f*x+%.4f" % (reg.coef_,reg.intercept_))
print("coef_ :  ",reg.coef_)
plt.scatter(x,y,label='x, y')#散点图
plt.plot(x,y_predict,label='predict')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('1')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#数据预处理 
x=np.arange(1, 41).reshape(-1, 1)
y=np.array(np.loadtxt('1.csv',delimiter=',',skiprows=1,usecols=3, max_rows=40))
y=y[::-1]
#线性回归
reg = LinearRegression()
reg.fit(x,y)
y_predict = reg.predict(x)
#绘图并输出方程、系数
print("Equation:","y=%.4f*x+%.4f" % (reg.coef_,reg.intercept_))
print("coef_ :  ",reg.coef_)

plt.plot(x, y)
plt.plot(x,y_predict)
plt.xlabel('Day')
plt.ylabel('Price')
plt.title("2")
plt.show()
```
